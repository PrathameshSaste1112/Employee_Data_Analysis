```python
catergorical_columns=["Career","Gender"]
for i in catergorical_columns:
  encoder=LabelEncoder()
  df[i]=encoder.fit_transform(df[i])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
for column in df.select_dtypes(include=['number']).columns:
       plt.figure(figsize=(8, 6))
       sns.histplot(df[column], bins=20)
       plt.title(f'Distribution of {column}')
       plt.xlabel(column)
       plt.ylabel('Frequency')
       plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

```python
for column in df.select_dtypes(include=['object']).columns:
       plt.figure(figsize=(8, 6))
       sns.countplot(x=column, data=df)
       plt.title(f'Distribution of {column}')
       plt.xlabel(column)
       plt.ylabel('Count')
       plt.xticks(rotation=45, ha='right')
       plt.show()
```

```python
for column in df.select_dtypes(include=['number']).columns:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x='Career', y=column, data=df)
        plt.title(f'{column} Distribution by Career')
        plt.xlabel('Career')
        plt.ylabel(column)
        plt.show()
```

```python
for column in df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x='Career', y=column, data=df)
    plt.title(f'{column} Distribution by Career')
    plt.xlabel('Career')
    plt.ylabel(column)
    plt.show()
```

```python
for column1 in df.select_dtypes(include=['number']).columns:
    for column2 in df.select_dtypes(include=['number']).columns:
        if column1 != column2:
            plt.figure(figsize=(8, 6))
            sns.jointplot(x=column1, y=column2, data=df, kind='scatter')  # kind='kde', 'reg', 'hex' for variations
            plt.show()
```
